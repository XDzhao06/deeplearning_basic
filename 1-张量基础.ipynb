# 张量：默认数据类型为 float32 (torch.FloatTensor)

torch.tensor 根据指定数据创建张量

torch.Tensor 除了小写能干的，还可以根据形状创建张量

```python
import torch
import numpy as np

def dm1():
    # 场景1：标量->张量
    t1 = torch.tensor(10)
    print(f't1 = {t1}, type:{type(t1)}')
    print('-' * 30)

    # 场景2：二维列表->张量
    data = [[1,2,3],[4,5,6]]
    t2 = torch.tensor(data)
    print(f't2 = {t2}, type:{type(t2)}')
    print('-' * 30)


    # 场景3：numpy数组->张量
    data = np.random.randint(0,10,size = (2,3))
    t3 = torch.tensor(data)
    print(f't3 = {t3}, type:{type(t3)}')
    print('-' * 30)

    # t4 = torch.tensor(data:2, dtype:3) 报错

def dm2():
    t1 = torch.Tensor(10)
    print(f't1 = {t1}, type:{type(t1)}')
    print('-' * 30)

    data = [[1,2,3],[4,5,6]]
    t2 = torch.Tensor(data)
    print(f't2 = {t2}, type:{type(t2)}')
    print('-' * 30)

    data = np.random.randint(0,10,size = (2,3))
    t3 = torch.Tensor(data)
    print(f't3 = {t3}, type:{type(t3)}')
    print('-' * 30)

    t4 = torch.Tensor(2,3)
    print(f't4 = {t4}, type:{type(t4)}')

def dm3():
    t1 = torch.IntTensor(10)
    print(f't1 = {t1}, type:{type(t1)}')
    print('-' * 30)

    data = [[1,2,3],[4,5,6]]
    t2 = torch.IntTensor(data)
    print(f't2 = {t2}, type:{type(t2)}')
    print('-' * 30)

    data = np.random.randint(0,10,size = (2,3))
    t3 = torch.IntTensor(data)
    print(f't3 = {t3}, type:{type(t3)}')
    print('-' * 30)

    # 场景4：如果类型不匹配，会自动尝试转换类型
    data = np.random.randint(0,10,size = (2,3))
    t4 = torch.FloatTensor(data)
    print(f't4 = {t4}, type:{type(t4)}')
    print('-' * 30)

if __name__ == '__main__':
    dm1()
    #dm2()
    #dm3()
```

# 创建线性和随机张量

```python
def dm4():
    # 场景1：指定范围的线性张量
    t1 = torch.arange(0,10,2) # 起始，结束，步长
    print(f't1 = {t1}, type:{type(t1)}')
    print('-' * 30)

    # 场景2：创建指定范围的等差数列张量
    t2 = torch.linspace(1,10,4) # 起始，结束，元素个数
    print(f't2 = {t2}, type:{type(t2)}')
    print('-' * 30)


def dm5():
    # 设置随机种子
    #torch.initial_seed() # 采用系统当前时间戳作为随机种子
    torch.manual_seed(42) # 依据给定种子作为随机种子

    # 创建随机张量
    # 场景1：均匀分布的(0,1)随机张量
    t1 = torch.rand(size = (2,3))
    print(f't1 = {t1}, type:{type(t1)}')
    print('-' * 30)

    # 场景2：正态分布的
    t2 = torch.randn(size = (2,3))
    print(f't2 = {t2}, type:{type(t2)}')
    print('-' * 30)

    # 场景3：整数的
    t3 = torch.randint(0,10,size = (2,3))
    print(f't3 = {t3}, type:{type(t3)}')
    print('-' * 30)

if __name__ == '__main__':
    dm4()
    dm5()
```

# 创建0、1、指定值张量

```python
def dm6():
    # 全1张量
    t1 = torch.ones(2,3)
    print(f't1 = {t1}, type:{type(t1)}')
    print('-' * 30)

    t2 = torch.tensor([[1,2],[3,4],[5,6]])
    print(f't2 = {t2}, type:{type(t2)}')
    t3 = torch.ones_like(t2)
    print(f't3 = {t3}, type:{type(t3)}')
    print('-' * 30)

    # 全0张量
    t1 = torch.zeros(2,3)
    print(f't1 = {t1}, type:{type(t1)}')
    print('-' * 30)

    t2 = torch.tensor([[1,2],[3,4],[5,6]])
    print(f't2 = {t2}, type:{type(t2)}')
    t3 = torch.zeros_like(t2)
    print(f't3 = {t3}, type:{type(t3)}')
    print('-' * 30)
    
    # 全指定值张量
    t1 = torch.full(size = (2, 3), fill_value = 255)
    print(f't1 = {t1}, type:{type(t1)}')
    print('-' * 30)

    t2 = torch.tensor([[1,2],[3,4],[5,6]])
    print(f't2 = {t2}, type:{type(t2)}')
    t3 = torch.full_like(t2, fill_value = 255)
    print(f't3 = {t3}, type:{type(t3)}')
    print('-' * 30)


if __name__ == '__main__':
        dm6()
```

# 创建指定类型张量

```python
# 场景1：直接创建
t1 = torch.tensor([1,2,3,4,5], dtype = torch.float)
print(f't1 = {t1}, type:{type(t1)}, 元素type:{t1.dtype}')
print('-' * 30)

# 场景2：类型转换
t2 = t1.type(torch.int16) #方案1
print(f't2 = {t2}, type:{type(t2)}, 元素type:{t2.dtype}')
print('-' * 30)

print(t2.half()) #方案2
print(t2.double())
```
